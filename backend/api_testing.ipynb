## Public endpoint-—ã

```python
import requests

BASE_URL = "https://it-hackathon-team02.mephi.ru/api/public"

def check_not_empty(resp):
    resp.raise_for_status()
    data = resp.json() if resp.headers.get("content-type","").startswith("application/json") else resp.content
    assert data, f"–û—Ç–≤–µ—Ç –ø—É—Å—Ç–æ–π: {resp.url}"
    return data

def test_news_list():
    data = check_not_empty(requests.get(f"{BASE_URL}/news"))
    print("news:", len(data))

def test_news_by_id(news_id=1):
    data = check_not_empty(requests.get(f"{BASE_URL}/news/{news_id}"))
    print("news by id:", data["id"])

def test_events():
    data = check_not_empty(requests.get(f"{BASE_URL}/events"))
    print("events:", len(data))

def test_event_by_id(event_id=1):
    data = check_not_empty(requests.get(f"{BASE_URL}/events/{event_id}"))
    print("event by id:", data["id"])

def test_nkos():
    data = check_not_empty(requests.get(f"{BASE_URL}/nkos"))
    print("nkos:", len(data))

def test_nko_by_id(nko_id=1):
    data = check_not_empty(requests.get(f"{BASE_URL}/nkos/{nko_id}"))
    print("nko by id:", data["id"])

def test_news_categories():
    data = check_not_empty(requests.get(f"{BASE_URL}/categories/news"))
    print("news categories:", data)

def test_event_categories():
    data = check_not_empty(requests.get(f"{BASE_URL}/categories/events"))
    print("event categories:", data)

def test_nko_categories():
    data = check_not_empty(requests.get(f"{BASE_URL}/categories/nkos"))
    print("nko categories:", data)

def test_kb():
    data = check_not_empty(requests.get(f"{BASE_URL}/knowledge-base"))
    print("kb:", len(data))

def test_kb_by_id(kb_id=1):
    data = check_not_empty(requests.get(f"{BASE_URL}/knowledge-base/{kb_id}"))
    print("kb by id:", data["id"])

def test_file(file_path="files/news/application-form.docx"):
    resp = requests.get(f"{BASE_URL}/files", params={"file_path": file_path})
    resp.raise_for_status()
    assert resp.content, "–§–∞–π–ª –ø—É—Å—Ç–æ–π"
    print("file bytes:", len(resp.content))

def test_cities():
    """–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–∏—è –≤—Å–µ—Ö –≥–æ—Ä–æ–¥–æ–≤."""
    data = check_not_empty(requests.get(f"{BASE_URL}/cities"))
    print("cities:", len(data))
    if data:
        print(f"–ü—Ä–∏–º–µ—Ä –≥–æ—Ä–æ–¥–∞: {data[0].get('name')} (id: {data[0].get('id')})")

def test_cities_with_organizations():
    """–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–∏—è –≥–æ—Ä–æ–¥–æ–≤ —Å –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è–º–∏."""
    data = check_not_empty(requests.get(f"{BASE_URL}/cities/with-organizations"))
    print("cities with organizations:", len(data))
    if data:
        print(f"–ü—Ä–∏–º–µ—Ä –≥–æ—Ä–æ–¥–∞ —Å –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è–º–∏: {data[0].get('name')} (id: {data[0].get('id')})")
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å
        city = data[0]
        if city.get('lat') and city.get('long'):
            print(f"  –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã: lat={city.get('lat')}, long={city.get('long')}")

if __name__ == "__main__":
    test_news_list()
    test_news_by_id()
    test_events()
    test_event_by_id()
    test_nkos()
    test_nko_by_id()
    test_kb()
    test_kb_by_id()
    test_news_categories()
    test_event_categories()
    test_nko_categories()
    test_file()
    test_cities()
    test_cities_with_organizations()
```

## News endpoint-—ã

```python
import requests
from pprint import pprint
from datetime import datetime

BASE_URL = "https://it-hackathon-team02.mephi.ru/api"  # –ø–æ–ø—Ä–∞–≤—å—Ç–µ –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏

ADMIN_EMAIL = "1@example.com"
ADMIN_PASSWORD = "1"

import uuid
from pathlib import Path

UPLOAD_FILE_PATH = Path("test_data/news/test_upload.txt")
UPLOAD_TARGET = "attachments/test_upload.txt"
DELETE_TARGET = UPLOAD_TARGET  # –ø—É—Ç—å —Å–æ–≤–ø–∞–¥–∞–µ—Ç; –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –∑–∞–¥–∞—Ç—å –¥—Ä—É–≥–æ–π

UPLOAD_IMAGE_PATH = Path("test_data/news/test_image.jpg")
UPLOAD_IMAGE_TARGET = "images/test_image.jpg"
DELETE_IMAGE_TARGET = UPLOAD_IMAGE_TARGET

def get_token():
    resp = requests.post(
        f"{BASE_URL}/auth/login",
        json={"email": ADMIN_EMAIL, "password": ADMIN_PASSWORD},
        timeout=10,
    )
    resp.raise_for_status()
    token = resp.json()["access_token"]
    return {"Authorization": f"Bearer {token}"}


def create_news(headers):
    payload = {
        "title": "–¢–µ—Å—Ç–æ–≤–∞—è –Ω–æ–≤–æ—Å—Ç—å",
        "short_description": "–ö—Ä–∞—Ç–∫–æ –æ —Ç–µ—Å—Ç–µ",
        "full_description": "–ü–æ–ª–Ω—ã–π —Ç–µ–∫—Å—Ç —Ç–µ—Å—Ç–æ–≤–æ–π –Ω–æ–≤–æ—Å—Ç–∏",
        "date_event": datetime.utcnow().isoformat(),
        "category_news_name": "–ê–≤—Ç–æ—Ç–µ—Å—Ç—ã",
        "city_name": "–¢–µ—Å—Ç–æ–≥—Ä–∞–¥",
        "hashtags": ["—Ç–µ—Å—Ç", "api"],
        "images": ["files/news/presentation-center.pdf"],
        "files": ["files/news/event-plan.docx"],
    }
    resp = requests.post(f"{BASE_URL}/admin_news", json=payload, headers=headers, timeout=10)
    resp.raise_for_status()
    return resp.json()


def update_news(news_id, headers):
    payload = {
        "title": "–¢–µ—Å—Ç–æ–≤–∞—è –Ω–æ–≤–æ—Å—Ç—å (–æ–±–Ω–æ–≤–ª–µ–Ω–æ)",
        "short_description": "–û–±–Ω–æ–≤–ª—ë–Ω–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ",
        "hashtags": ["—Ç–µ—Å—Ç", "api", "update"],
        "images": [],
        "files": [],
    }
    resp = requests.put(
        f"{BASE_URL}/admin_news/{news_id}",
        json=payload,
        headers=headers,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def delete_news(news_id, headers):
    resp = requests.delete(
        f"{BASE_URL}/admin_news/{news_id}",
        headers=headers,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def upload_news_file(news_id, headers, file_path=UPLOAD_FILE_PATH):
    file_path.parent.mkdir(parents=True, exist_ok=True)
    file_path.write_text(f"Test content {uuid.uuid4()}")

    with open(file_path, "rb") as fh:
        files = {
            "file": (file_path.name, fh, "text/plain"),
        }
        data = {
            "news_id": str(news_id),
        }
        resp = requests.post(
            f"{BASE_URL}/admin_news/upload-file",
            headers=headers,
            data=data,
            files=files,
            timeout=10,
        )
        resp.raise_for_status()
        return resp.json()


def delete_news_file(news_id, headers, path):
    payload = {
        "news_id": news_id,
        "file_path": path,
    }
    resp = requests.delete(
        f"{BASE_URL}/admin_news/delete-file",
        headers=headers,
        json=payload,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def upload_news_image(news_id, headers, image_path=UPLOAD_IMAGE_PATH):
    image_path.parent.mkdir(parents=True, exist_ok=True)
    image_path.write_bytes(b"dummy image data")

    with open(image_path, "rb") as fh:
        files = {
            "image": (image_path.name, fh, "image/jpeg"),
        }
        data = {
            "news_id": str(news_id),
        }
        resp = requests.post(
            f"{BASE_URL}/admin_news/upload-image",
            headers=headers,
            data=data,
            files=files,
            timeout=10,
        )
        resp.raise_for_status()
        return resp.json()


def delete_news_image(news_id, headers, path):
    payload = {
        "news_id": news_id,
        "image_path": path,
    }
    resp = requests.delete(
        f"{BASE_URL}/admin_news/delete-image",
        headers=headers,
        json=payload,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def main():
    headers = get_token()

    print("\n–°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    created = create_news(headers)
    pprint(created)
    news_id = created["id"]

    print("\n–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ –∫ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    uploaded = upload_news_file(news_id, headers)
    pprint(uploaded)

    print("\n–£–¥–∞–ª–µ–Ω–∏–µ —Ñ–∞–π–ª–∞ –∏–∑ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    deleted_file = delete_news_file(news_id, headers, path=uploaded["path"])
    pprint(deleted_file)

    print("\n–ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∫ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    uploaded_image = upload_news_image(news_id, headers)
    pprint(uploaded_image)

    print("\n–£–¥–∞–ª–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏–∑ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    deleted_image = delete_news_image(news_id, headers, path=uploaded_image["path"])
    pprint(deleted_image)

    print("\n–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    updated = update_news(news_id, headers)
    pprint(updated)

    print("\n–£–¥–∞–ª–µ–Ω–∏–µ –Ω–æ–≤–æ—Å—Ç–∏‚Ä¶")
    deleted = delete_news(news_id, headers)
    pprint(deleted)

if __name__ == "__main__":
    main()
```

## –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ endpoint-–æ–≤ –∞–¥–º–∏–Ω–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏—è —Å–æ–±—ã—Ç–∏–π

```python
import io
import os
import random
import string
import requests
from pprint import pprint

BASE_URL = "https://it-hackathon-team02.mephi.ru/api"
ADMIN_EMAIL = "1@example.com"
ADMIN_PASSWORD = "1"

ORG_ID = 1                 # –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è, —Å –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç–µ —Ä–∞–±–æ—Ç–∞—Ç—å
TEST_IMAGE = "test_data/event/event_test.jpg"  # –∑–∞—Ä–∞–Ω–µ–µ –ø–æ–ª–æ–∂–∏—Ç–µ —Ñ–∞–π–ª-–æ–±—Ä–∞–∑–µ—Ü

def auth_headers():
    resp = requests.post(
        f"{BASE_URL}/auth/login",
        json={"email": ADMIN_EMAIL, "password": ADMIN_PASSWORD},
        timeout=10,
    )
    resp.raise_for_status()
    token = resp.json()["access_token"]
    return {"Authorization": f"Bearer {token}"}

def random_title(prefix="Auto Event"):
    suffix = "".join(random.choices(string.ascii_uppercase + string.digits, k=4))
    return f"{prefix} {suffix}"

def create_event(headers):
    payload = {
        "title": random_title(),
        "description": "–ö–æ—Ä–æ—Ç–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ",
        "full_description": "–ü–æ–ª–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è",
        "organization_id": ORG_ID,
        "type_event_name": "–í–æ–ª–æ–Ω—Ç–µ—Ä—Å—Ç–≤–æ",
        "category_event_name": "–≠–∫–æ–ª–æ–≥–∏—è",
        "quantity_participant": 25,
    }
    resp = requests.post(f"{BASE_URL}/admin/events", json=payload, headers=headers, timeout=10)
    resp.raise_for_status()
    data = resp.json()
    print("\n[create]"); pprint(data)
    return data

def update_event(event_id, headers):
    payload = {
        "title": random_title("Updated Event"),
        "quantity_participant": 50,
        "images": ["files/events/images/demo_placeholder.jpg"],
    }
    resp = requests.put(f"{BASE_URL}/admin/events/{event_id}", json=payload, headers=headers, timeout=10)
    resp.raise_for_status()
    print("\n[update]"); pprint(resp.json())

def upload_image(event_id, headers, filename):
    if not os.path.exists(filename):
        raise FileNotFoundError(f"–¢–µ—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª {filename} –Ω–µ –Ω–∞–π–¥–µ–Ω")
    with open(filename, "rb") as fh:
        files = {"image": (os.path.basename(filename), fh, "image/jpeg")}
        data = {"event_id": str(event_id)}
        resp = requests.post(
            f"{BASE_URL}/admin/events/upload-image",
            data=data,
            files=files,
            headers=headers,
            timeout=10,
        )
    resp.raise_for_status()
    print("\n[upload-image]"); pprint(resp.json())
    return resp.json()["path"]

def delete_image(event_id, image_path, headers):
    payload = {"event_id": event_id, "image_path": image_path}
    resp = requests.delete(f"{BASE_URL}/admin/events/delete-image", json=payload, headers=headers, timeout=10)
    resp.raise_for_status()
    print("\n[delete-image]"); pprint(resp.json())

def list_org_events(org_id, headers):
    resp = requests.get(f"{BASE_URL}/admin/events/organization/{org_id}", headers=headers, timeout=10)
    resp.raise_for_status()
    print("\n[list-org] –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ:", len(resp.json()))

def delete_event(event_id, headers):
    resp = requests.delete(f"{BASE_URL}/admin/events/{event_id}", headers=headers, timeout=10)
    resp.raise_for_status()
    print("\n[delete-event]"); pprint(resp.json())

def main():
    headers = auth_headers()
    created = create_event(headers)
    event_id = created["id"]
    print(event_id)
    print(type(event_id))
    update_event(event_id, headers)
    uploaded_path = upload_image(event_id, headers, TEST_IMAGE)
    print(uploaded_path)
    print(type(uploaded_path))
    delete_image(event_id, uploaded_path, headers)
    list_org_events(ORG_ID, headers)
    delete_event(event_id, headers)

if __name__ == "__main__":
    main()
```

```python
import requests
from pprint import pprint
from datetime import datetime
import uuid
from pathlib import Path

BASE_URL = "https://it-hackathon-team02.mephi.ru/api"

ADMIN_EMAIL = "1@example.com"
ADMIN_PASSWORD = "1"

UPLOAD_FILE_PATH = Path("test_data/knowledge_base/test_upload.txt")


def get_token():
    """–ü–æ–ª—É—á–µ–Ω–∏–µ —Ç–æ–∫–µ–Ω–∞ –∞–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏."""
    resp = requests.post(
        f"{BASE_URL}/auth/login",
        json={"email": ADMIN_EMAIL, "password": ADMIN_PASSWORD},
        timeout=10,
    )
    resp.raise_for_status()
    token = resp.json()["access_token"]
    return {"Authorization": f"Bearer {token}"}


def create_knowledge_base(headers):
    """–°–æ–∑–¥–∞–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π."""
    payload = {
        "title": "–¢–µ—Å—Ç–æ–≤—ã–π –º–∞—Ç–µ—Ä–∏–∞–ª –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π",
        "description": "–ö—Ä–∞—Ç–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞",
        "content": "–ü–æ–ª–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π",
        "video_url": "https://example.com/video.mp4",
        "category_knowledge_base_name": "–ê–≤—Ç–æ—Ç–µ—Å—Ç—ã",
        "type_material_name": "–î–æ–∫—É–º–µ–Ω—Ç",
        "files": [],
    }
    resp = requests.post(
        f"{BASE_URL}/admin_knowledge_base",
        json=payload,
        headers=headers,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def update_knowledge_base(kb_id, headers):
    """–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π."""
    payload = {
        "title": "–¢–µ—Å—Ç–æ–≤—ã–π –º–∞—Ç–µ—Ä–∏–∞–ª –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π (–æ–±–Ω–æ–≤–ª–µ–Ω–æ)",
        "description": "–û–±–Ω–æ–≤–ª—ë–Ω–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ",
        "content": "–û–±–Ω–æ–≤–ª—ë–Ω–Ω–æ–µ –ø–æ–ª–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ",
        "files": [],
    }
    resp = requests.put(
        f"{BASE_URL}/admin_knowledge_base/{kb_id}",
        json=payload,
        headers=headers,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def delete_knowledge_base(kb_id, headers):
    """–£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π."""
    resp = requests.delete(
        f"{BASE_URL}/admin_knowledge_base/{kb_id}",
        headers=headers,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def upload_knowledge_base_file(kb_id, headers, file_path=UPLOAD_FILE_PATH):
    """–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ –¥–ª—è —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π."""
    file_path.parent.mkdir(parents=True, exist_ok=True)
    file_path.write_text(f"Test content {uuid.uuid4()}")

    with open(file_path, "rb") as fh:
        files = {
            "file": (file_path.name, fh, "text/plain"),
        }
        data = {
            "knowledge_base_id": str(kb_id),
        }
        resp = requests.post(
            f"{BASE_URL}/admin_knowledge_base/upload-file",
            headers=headers,
            data=data,
            files=files,
            timeout=10,
        )
        resp.raise_for_status()
        return resp.json()


def delete_knowledge_base_file(kb_id, headers, path):
    """–£–¥–∞–ª–µ–Ω–∏–µ —Ñ–∞–π–ª–∞ –∏–∑ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π."""
    payload = {
        "knowledge_base_id": kb_id,
        "file_path": path,
    }
    resp = requests.delete(
        f"{BASE_URL}/admin_knowledge_base/delete-file",
        headers=headers,
        # json=payload,
        data=payload,
        timeout=10,
    )
    resp.raise_for_status()
    return resp.json()


def main():
    """–û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–ø—É—Å–∫–∞ –≤—Å–µ—Ö —Ç–µ—Å—Ç–æ–≤."""
    headers = get_token()

    print("\n" + "="*60)
    print("–°–æ–∑–¥–∞–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π‚Ä¶")
    print("="*60)
    created = create_knowledge_base(headers)
    pprint(created)
    kb_id = created["id"]

    print("\n" + "="*60)
    print("–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ –∫ —ç–ª–µ–º–µ–Ω—Ç—É –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π‚Ä¶")
    print("="*60)
    uploaded = upload_knowledge_base_file(kb_id, headers)
    pprint(uploaded)

    print("\n" + "="*60)
    print("–£–¥–∞–ª–µ–Ω–∏–µ —Ñ–∞–π–ª–∞ –∏–∑ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π‚Ä¶")
    print("="*60)
    deleted_file = delete_knowledge_base_file(kb_id, headers, path=uploaded["path"])
    pprint(deleted_file)

    print("\n" + "="*60)
    print("–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π‚Ä¶")
    print("="*60)
    updated = update_knowledge_base(kb_id, headers)
    pprint(updated)

    print("\n" + "="*60)
    print("–£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π‚Ä¶")
    print("="*60)
    deleted = delete_knowledge_base(kb_id, headers)
    pprint(deleted)

    print("\n" + "="*60)
    print("‚úì –í—Å–µ —Ç–µ—Å—Ç—ã –∑–∞–≤–µ—Ä—à–µ–Ω—ã")
    print("="*60)


if __name__ == "__main__":
    main()
```
