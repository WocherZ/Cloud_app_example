```python
import pandas as pd
```

```python
# Прочитаем все листы Excel с отступом в 1 строку и 1 колонку
excel_file = pd.ExcelFile('Мастер таблица для НКО и КСО.xlsx')
dfs = {}
for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet_name, skiprows=1, usecols=lambda x: True)[1:]  # Отступ 1 строка и 1 колонка
    dfs[sheet_name] = df
dfs  # Словарь датафреймов по листам
```
